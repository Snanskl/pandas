**7. On which days of the week do most crimes occur?**

```python
#Assigning 'Date Rptd' to datetime format
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], errors='coerce')

#day_name() is a function that will return the day of the week 
#'Date Rptd' is now being assign to 'Day of Week' to turn into a string format
df['Day of Week'] = df['Date Rptd'].dt.day_name()

#Count the number of crimes that happened on each day of the week
day_counts = df['Day of Week'].value_counts()

most_crimes_day = day_counts.head(7)
print(most_crimes_day)
```
