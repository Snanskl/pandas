# Working with the Netflix csv file

```python
import pandas as pd
import numpy as np
```

1. Read the file into pandas 

```python
df = pd.read_csv('C:\\Users\\ployp\\OneDrive\\Desktop\\panda\\MoviesReview\\netflix.csv') 
```

2. Read the data

```python
df
df.info()

#From this we can see that the data is not clean and we need to clean it up.
#The first thing to do is to remove any unnessary string from column 'show_id' and 'date_added'
#With the df.info() we can also see that majority of columns have a datatype as object, we need to change this to the correct datatype
```

3. Clean the column 'show_id' and 'date_added'

```python
#Clean up the 'show_id' column

#cleaning up the column by replacing the 's' with an empty string
df['show_id'] = df['show_id'].str.replace('s', '')

df
```

```python
#Clean up the 'date_added' column

#Initially, i tried to use just "pd.to_datetime(df['date_added'], format = '%B %d, %Y')" but it gave me an error because of the white spaces in the column
#So i used the .strip() method to remove the white spaces first
df['date_added'] = df['date_added'].str.strip()

#Now we can convert the column to datetime
df['date_added'] = pd.to_datetime(df['date_added'], format = '%B %d, %Y')

df
```

4. Fixing the 'duration' column

```python
#Fixing the duration column so that it only contains the number of minutes


#ValueError: Cannot mask with non-boolean array containing NA / NaN values
#seasons_only = df['duration'][df['duration'].str.contains('Season')]

#In this syntax it reads take the dataframe column name 'duration' and column values that contain 'Season' and store it in the variable 'seasons_only' where the column values are not null
season_only = df['duration'][df['duration'].str.contains('Season', na=False)]

season_only.value_counts()
```

```python
#Make the column 'duration' only contain the number of minutes instead of 'Seasons' 
df = df[~df['duration'].str.contains('Season|Seasons', na=False)]

df
```

5. Place an index

```python
#inplace mean that i want to make a chnage to be permanent
df.set_index('show_id', inplace=True)
```

6. Chnaging all the column to the correct datatypes

```python
#We use .loc to select the columns we want to change the datatype, and with ':' specifies for all rows, and then also specify the name of the column
df.loc[:, 'date_added'] = pd.to_datetime(df['date_added'], format = '%B %d, %Y')
```

```python
#When using errors='coerce' it will ignore any errors and set the value to NaN instead of raising errors
df.loc[:, 'release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

```python
#We use the .str.replace() method to remove the 'min' string from the column
df.loc[:, 'duration'] = df['duration'].str.replace(' min', '')
```

```python
#Make sure that the column 'duration_min' is of type int
df.loc[:, 'duration'] = pd.to_numeric(df['duration'], errors='coerce')
```

```python
#change the column name from 'duration' to 'duration_min'
df.rename(columns={'duration' : 'duration_min'}, inplace=True)
```
